```python
import pandas as pd
invivo_hek = pd.read_csv("hek_invivo_large.csv")
```

```python
# invivo_hek= invivo_hek[invivo_hek["Kinase Classification"].notna()]
invivo_hek_sorted = invivo_hek.sort_values(by="Kinase Classification")

invivo_hek_sorted = invivo_hek_sorted[invivo_hek_sorted["ProteinDescriptions"].str.contains("kinase", case=False, na=False)]
invivo_hek_sorted = invivo_hek_sorted.reset_index(drop=True)
invivo_hek_sorted = invivo_hek_sorted.drop(columns=["HGNC"])

invivo_hek_sorted.to_csv("invivo_hek_sorted.csv")

```
